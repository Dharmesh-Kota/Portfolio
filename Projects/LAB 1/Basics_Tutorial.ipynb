## Importing Library

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(45)
```

## Matplotlib Tutorial

```python
N = 100 # Sample size
X = np.linspace(0.0, 20.0, N)
Y_1 = np.sin(X)
Y_2 = np.cos(X)
```

```python
plt.plot(X, Y_1)
plt.show()
```

```python
plt.clf()
plt.plot(X, Y_1)
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.show()
```

```python
plt.clf()
plt.plot(X, Y_1, c='g')
plt.plot(X, Y_2, c='k')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.show()
```

```python
Y_cap = np.linspace(-1.0, 1.0, N) # Prediction

plt.clf()
plt.plot(X, Y_1, 'g*')
plt.plot(X, Y_cap, c='r')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Line Plot')
plt.show()
```

```python
plt.clf()
plt.scatter(X, Y_1, marker='^', c='g')
plt.plot(X, Y_cap, c='r')
plt.xlabel('X Axis Title')
plt.ylabel('Y Axis Title')
plt.title('Scatter Plot')
plt.show()
```

### A Simple Regression Model

```python
# A simple regression model
X = np.linspace(0.0, 20.0, N)
Y = X + np.random.randn(N,)
plt.clf()
plt.scatter(X, Y, marker='*', c='g')
plt.xlabel('Y Prediction')
plt.ylabel('X Data')
plt.title('Regression')
plt.show()
```

```python
total_epochs = 10

for epoch in range(total_epochs):
    theta = np.random.randn(1,)
    bias  = np.random.randn(1,)
    Y_cap = theta * X + bias
    plt.clf()
    plt.scatter(X, Y, marker='*', c='g')
    plt.plot(X, Y_cap, c='r')
    plt.xlabel('Y Prediction')
    plt.ylabel('X Data')
    plt.title('Regression')
    plt.show()
```

```python
os.makedirs('output')
total_epochs = 10

for epoch in range(total_epochs):
    theta = np.random.randn(1,)
    bias  = np.random.randn(1,)
    Y_cap = theta * X + bias
    plt.clf()
    plt.scatter(X, Y, marker='*', c='g')
    plt.plot(X, Y_cap, c='r')
    plt.xlabel('Y Prediction')
    plt.ylabel('X Data')
    plt.title('Regression')
    plt.savefig(f'output/{epoch}.png', bbox_inches='tight', dpi=300)
```

### Creating GIF File

```python
from PIL import Image
from glob import glob
from natsort import natsorted

os.makedirs('results')

def create_gif(image_paths, output_gif_path, duration=500):
    images = [Image.open(image_path) for image_path in image_paths]
    # Save as GIF
    images[0].save(
                output_gif_path,
                save_all=True,
                append_images=images[1:],
                duration=duration,
                loop=0 # 0 means infinite loop
                )

create_gif(image_paths = natsorted(glob('output/*')), output_gif_path='results/gradient_update.gif')
```
